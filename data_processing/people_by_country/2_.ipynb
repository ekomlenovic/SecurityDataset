```python
import pandas as pd
import json
pd.set_option('display.max_columns', None)
```

```python
# Load your datasets
data = pd.read_excel('../../data/globalterrorismdb_0522dist.xlsx')
data_ppl_year = pd.read_csv('../../data/processed/ppl_country_years.csv')
with open('corrections.json', 'r') as file:
    corrections = json.load(file)
data['country_txt'] = data['country_txt'].replace(corrections)
```

```python
data.head()
```

```python
data_ppl_year.head()
```

```python
data_filtered = data[data['iyear'] >= 1970]

attacks_per_year_country = data_filtered.groupby(['iyear', 'country_txt']).size().reset_index(name='num_attacks')
attacks_per_year_country
```

```python
data_ppl_year_melted = data_ppl_year.melt(id_vars=['Country Name'], var_name='iyear', value_name='population')
data_ppl_year_melted['iyear'] = data_ppl_year_melted['iyear'].astype(int)
data_ppl_year_melted.rename(columns={'Country Name': 'country_txt'}, inplace=True)
data_ppl_year_melted
```

```python
merged_data = pd.merge(attacks_per_year_country, data_ppl_year_melted, on=['iyear', 'country_txt'], how='left')
merged_data['attacks_per_population'] = merged_data['num_attacks'] / merged_data['population']
merged_data
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


filtered_data = merged_data

plt.figure(figsize=(12, 8))

sns.lineplot(data=filtered_data, x='iyear', y='attacks_per_population', hue='country_txt', marker="o")

plt.title('Attacks per Population by Year Countries', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Attacks per Population', fontsize=14)

plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()

plt.show()
```
